```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Load a toy dataset about diabetes patients. Dataset documented [here](https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset).

X, y are conventional names for the input data and the output prediction (target). (I don't know _whose_ convention...)

```python
X, y = datasets.load_diabetes(return_X_y=True)

print(f"Shape of data matrix: {X.shape}")
print(f"Number of patients in dataset: {len(X)}")

print(f"Sample entry: {X[0]}")
```

```python
# Select portion of dataset for finding relationships between.
#
# We're using numpy array slicing. Turns the dataset into a column vector (with
# np.newaxis) containing only one dimension of the original data (chosen by the
# integer).
X_narrowed = X[:, np.newaxis, 0]

print(X_narrowed.shape)
```

```python
# Split up dataset into training and testing data.
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_narrowed, y, test_size=0.33)
```

```python
# Train a linear model on it.
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Create a prediction from the model.
y_pred = model.predict(X_test)
```

```python
# Plot the prediction over the actual test data.
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled age')
plt.ylabel('Disease Progression')
plt.title('Diabetes progression against age')
plt.show()
```
